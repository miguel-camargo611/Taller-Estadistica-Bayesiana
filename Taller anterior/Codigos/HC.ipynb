```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.cluster.hierarchy import fcluster

# Generar un conjunto de datos de muestra
np.random.seed(42)
X = np.random.multivariate_normal([10, 0], [[3, 1], [1, 4]], size=[50,])
Y = np.random.multivariate_normal([0, 20], [[3, 1], [1, 4]], size=[50,])
Z = np.random.multivariate_normal([20, 20], [[3, 1], [1, 4]], size=[50,])
data = np.concatenate((X, Y, Z),)

# Crear enlaces para cada método
methods = ['single', 'complete', 'average']
titles = ['Dendrograma con Enlace Simple', 'Dendrograma con Enlace Completo', 'Dendrograma con Enlace Promedio']

# Crear una figura con 3 subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for i, method in enumerate(methods):
    # Generar la matriz de enlace
    linked_data = linkage(data, method=method)
    
    # Color threshold para determinar el color de los clusters
    max_d = 7.08 if method == 'average' else None
    
    # Crear el dendrograma
    dendro = dendrogram(linked_data, ax=axes[i], color_threshold=max_d,
                        above_threshold_color='grey', orientation='top')
    
    # Asignar colores a los clusters
    clusters = fcluster(linked_data, 2, criterion='maxclust')
    for j, d in enumerate(data):
        axes[i].plot(d[0], d[1], color=plt.cm.nipy_spectral(clusters[j] / 10.))
    
    # Añadir título y quitar los ejes x e y
    axes[i].set_title(titles[i])
    axes[i].set_xticks([])
    axes[i].set_yticks([])

# Mostrar los dendrogramas
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.cluster.hierarchy import fcluster

# Generar un conjunto de datos de muestra
np.random.seed(42)
X = np.random.multivariate_normal([10, 0], [[3, 1], [1, 4]], size=[50,])
Y = np.random.multivariate_normal([0, 20], [[3, 1], [1, 4]], size=[50,])
data = np.concatenate((X, Y),)

# Generar la matriz de enlace con el método de enlace promedio
linked_data = linkage(data, method='average')

# Crear el dendrograma y asignar colores automáticamente
plt.figure(figsize=(10, 7))
dendrogram(linked_data,
           color_threshold=1.5, # Este valor determina la "altura" donde se corta el dendrograma para definir los colores
           above_threshold_color='grey')

plt.title('Dendrograma con Enlace Promedio')
plt.show()
```
