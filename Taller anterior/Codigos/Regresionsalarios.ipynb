## Métodos de Aprendizaje

El aprendizaje automático es un conjunto de herramientas que, a grandes rasgos, nos permiten “enseñar” a los ordenadores cómo realizar tareas proporcionando ejemplos de cómo deben realizarse. Por ejemplo, supongamos que deseamos escribir un programa para distinguir entre mensajes de correo electrónico válidos y spam no deseado. Podríamos intentar escribir un conjunto de reglas simples, por ejemplo, marcar mensajes que contengan ciertas características (como la palabra "viagra" o encabezados obviamente falsos). Sin embargo, escribir reglas para distinguir con precisión qué texto es válido en realidad puede ser bastante difícil de hacer bien, lo que da como resultado muchos mensajes de spam perdidos o, peor aún, muchos correos electrónicos perdidos.


Tipos de aprendizaje automático
Algunos de los principales tipos de aprendizaje automático son:

1. Aprendizaje supervisado, en el que los datos de entrenamiento se etiquetan con las respuestas correctas, por ejemplo, "correo no deseado" o "jamón". Los dos tipos más comunes de aprendizaje supervisado son la clasificación (donde los resultados son etiquetas discretas, como en el filtrado de spam) y la regresión (donde los resultados tienen un valor real).

2. Aprendizaje no supervisado, en el que se entrega una colección de datos etiquetados, que deseamos analizar y descubrir patrones en su interior. Los dos ejemplos más importantes son la reducción de dimensiones y el agrupamiento.

3. Aprendizaje por refuerzo, en el que un agente (p. ej., un robot o un controlador) busca aprender las acciones óptimas a realizar basándose en los resultados de acciones pasadas.


El aprendizaje automático brinda una amplia selección de opciones para responder estas preguntas, junto con la vasta experiencia de la comunidad en cuanto a qué métodos tienden a tener éxito en una clase particular de conjunto de datos. Algunos métodos más avanzados brindan formas de automatizar algunas de estas opciones, como la selección automática entre modelos alternativos, y existe una hermosa teoría que ayuda a obtener una comprensión más profunda del aprendizaje. En la práctica, no existe una única “bala de plata” para todo el aprendizaje. El uso del aprendizaje automático en la práctica requiere que haga uso de su propio conocimiento previo y experimentación para resolver problemas.

**Hold out**

El método de retención (Hold out) para entrenar los modelos de aprendizaje automático es una técnica que implica dividir los datos en diferentes conjuntos: un conjunto para entrenamiento y otros conjuntos para validación y prueba. El método de retención se utiliza para verificar qué tan bien funcionará un modelo de aprendizaje automático con los nuevos datos.

Una gran cantidad de datos permite hacer esta metodolgía en donde  principalmente se dividen los datos en:

* Datos de entrenamiento.

* Datos de verificación.

## Principal preocupación

Cada vez que discutimos la predicción del modelo, es importante comprender los errores de predicción (sesgo y varianza). Existe una compensación entre la capacidad de un modelo para minimizar el sesgo y la varianza. Obtener una comprensión adecuada de estos errores nos ayudaría no solo a construir modelos precisos, sino también a evitar el error de sobreajuste y ajuste insuficiente.
Entonces, comencemos con lo básico y veamos cómo marcan la diferencia en nuestros modelos de aprendizaje automático.

El sesgo es la diferencia entre la predicción promedio de nuestro modelo y el valor correcto que estamos tratando de predecir. El modelo con alto sesgo presta muy poca atención a los datos de entrenamiento y simplifica demasiado el modelo. Siempre conduce a un alto error en los datos de entrenamiento y prueba.

Ejemplo:

$$Y=f(x)+e$$

Donde una variable dependiente se predice por una función $f(x)$ que contiene variables independiente y un error $\epsilon$.



La varianza es la variabilidad de la predicción del modelo para un punto de datos dado o un valor que nos indica la dispersión de nuestros datos. El modelo con alta varianza presta mucha atención a los datos de entrenamiento y no generaliza sobre los datos que no ha visto antes. Como resultado, estos modelos funcionan muy bien con los datos de entrenamiento, pero tienen altas tasas de error con los datos de prueba.

El sesgo y la varianza forman el error y muestran en que medida se alejan las predicciones de las observaciones reales.

Ese error al cuadrado se obtiene de restar las observaciones de las predicciones.

$$Error=E[(Y-\hat{Y})^2]$$

Donde $\hat{Y}$ es producto de una función , por lo tanto el error se puede rescribir de esta forma:

$$Error(x)=(E[\hat{Y}-Y])^2+E[(\hat{Y}-E[\hat{Y}])^2]+\sigma_{e}^{2}$$

Donde la primer parte de la derecha de la igualdad es el sesgo y la siguiente parte es la varianza más un error que no se puede controlar.

En el aprendizaje supervisado, el subajuste  ocurre cuando un modelo no puede capturar el patrón subyacente de los datos. Estos modelos suelen tener un alto sesgo y una baja varianza. Ocurre cuando tenemos muy poca cantidad de datos para construir un modelo preciso o cuando intentamos construir un modelo lineal con datos no lineales. Además, este tipo de modelos son muy simples para capturar los patrones complejos en datos como la regresión lineal y logística.

En el aprendizaje supervisado, el sobreajuste ocurre cuando nuestro modelo captura el ruido junto con el patrón subyacente en los datos. Ocurre cuando entrenamos mucho nuestro modelo sobre un conjunto de datos ruidoso. Estos modelos tienen un sesgo bajo y una varianza alta. Estos modelos son muy complejos, como los árboles de decisión, que tienden a sobreajustarse.



## Gradiente descendiente

### Función de Perdida y Gradiente descendientes 

**Regresión Lineal**


$$h_\theta (x)=\theta_0 +\theta_1 x_1+ \theta_2 x_2$$

Donde $\theta_i$ son los parámetros del modelo, $x_i$ son los datos de entrada y $h_\theta(x)$ es la predicción del modelo.

$$h(x)=\sum_i=0^n \theta_i x_i= \theta^T x$$

Función de costo:

$$J(\theta)=\frac{1}{2m}\sum_{i=1}^m (h_\theta(x^{(i)})-y^{(i)})^2$$

Minimizar $J(\theta)$ es igual a encontrar el valor de $\theta$ que minimiza $J(\theta)$.Que inicia con $\theta$ y que actualiza $\theta$ con el valor que minimiza $J(\theta)$.

$$\theta_j = \theta_j - \alpha \frac{1}{m} \sum_{i=1}^m (h_\theta(x^{(i)})-y^{(i)})x_j^{(i)}$$

Donde $\alpha$ es la tasa de aprendizaje del gradiente.

$$ \frac{\partial}{\partial \theta_j} J(\theta) = \frac{\partial}{\partial \theta_j}\frac{1}{2} (h_\theta(x)-y)^2$$

$$ = 2.\frac{1}{2}(h_\theta(x)-y)\frac{\partial}{\partial \theta_j}(h_\theta(x)-y)$$

$$=(h_\theta (x)-y)x_j$$

**Para un entrenamiento simple**

$$ \theta_j = \theta_j - \alpha (h_\theta(y^{(i)})-h_\theta (x^{(i)})x_j^{(i)}$$

Esto se repite hasta llegar a la convergencia.

Cuanto debería ser el valor de $\alpha$?


## Metricas

En Regresión


La tasa de error de un modelo predictivo puede evaluarse aplicando varias métricas derivadas de funciones de perdida en ML y en estadística. El concepto básico de la evaluación de la precisión en el análisis de regresión consiste en comparar la variable objetivo con la predicha.
Recordamos como hallar algunas de las métricas más usadas:

+ $RMSE=\frac{1}{N}\sqrt{\sum_{i=1}^{N}(y_{i}-\hat{y}_{i})^2}$, Root Mean Squared Error

+ $R^{2}=1-\frac{\sum_{i=1}^{N}(y_{i}-\hat{y}_{i})^{2}}{\sum_{i=1}^{N}(y_{i}-\bar{y})^{2}}$, R-squared (Coefficient of determination) 

+ $MAPE=\frac{1}{N}\sum_{i=1}^{N}\left|\frac{y_{i}-\hat{y}_{i}}{y_{i}}\right|$, Mean absolute percentage error

```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# The data to fit
m = 20
theta1_true = 0.5
x = np.linspace(-1,1,m)
y = theta1_true * x

# The plot: LHS is the data, RHS will be the cost function.
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,6.15))
ax[0].scatter(x, y, marker='x', s=40, color='k')

def cost_func(theta1):
    """The cost function, J(theta1) describing the goodness of fit."""
    theta1 = np.atleast_2d(np.asarray(theta1))
    return np.average((y-hypothesis(x, theta1))**2, axis=1)/2

def hypothesis(x, theta1):
    """Our "hypothesis function", a straight line through the origin."""
    return theta1*x

# First construct a grid of theta1 parameter pairs and their corresponding
# cost function values.
theta1_grid = np.linspace(-0.2,1,50)
J_grid = cost_func(theta1_grid[:,np.newaxis])

# The cost function as a function of its single parameter, theta1.
ax[1].plot(theta1_grid, J_grid, 'k')

# Take N steps with learning rate alpha down the steepest gradient,
# starting at theta1 = 0.
N = 5
alpha = 1
theta1 = [0]
J = [cost_func(theta1[0])[0]]
for j in range(N-1):
    last_theta1 = theta1[-1]
    this_theta1 = last_theta1 - alpha / m * np.sum(
                                    (hypothesis(x, last_theta1) - y) * x)
    theta1.append(this_theta1)
    J.append(cost_func(this_theta1))

# Annotate the cost function plot with coloured points indicating the
# parameters chosen and red arrows indicating the steps down the gradient.
# Also plot the fit function on the LHS data plot in a matching colour.
colors = ['b', 'g', 'm', 'c', 'orange']
ax[0].plot(x, hypothesis(x, theta1[0]), color=colors[0], lw=2,
           label=r'$\theta_1 = {:.3f}$'.format(theta1[0]))
for j in range(1,N):
    ax[1].annotate('', xy=(theta1[j], J[j]), xytext=(theta1[j-1], J[j-1]),
                   arrowprops={'arrowstyle': '->', 'color': 'r', 'lw': 1},
                   va='center', ha='center')
    ax[0].plot(x, hypothesis(x, theta1[j]), color=colors[j], lw=2,
               label=r'$\theta_1 = {:.3f}$'.format(theta1[j]))

# Labels, titles and a legend.

ax[1].set_xlim(-0.2,1)
ax[1].set_xlabel(r'$\theta_1$')
ax[1].set_ylabel(r'$J(\theta_1)$')
ax[1].set_title('Cost function')
ax[0].set_xlabel(r'$x$')
ax[0].set_ylabel(r'$y$')
ax[0].set_title('Data and fit')
ax[0].legend(loc='upper left', fontsize='small')

plt.tight_layout()
plt.show()
```

```python
# Importar librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan
from scipy.stats import kstest
import scipy.stats as stats

# 🔹 Cargar datos desde CSV
df = pd.read_csv("wage.csv")


```

```python
df.head()

```

```python
# 🔹 Crear variable de experiencia al cuadrado
# 🔹 Crear nuevas variables necesarias
df["expersq"] = df["exper"] ** 2  # Experiencia al cuadrado
df["exper_female"] = df["exper"] * df["female"]  # Interacción experiencia * mujer


# 🔹 Definir variables predictoras y objetivo
X = df[["educ", "exper", "tenure", "female", "expersq", "exper_female"]]
y = df["lwage"]

# 🔹 Dividir en train y test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

```

```python
from sklearn.linear_model import SGDRegressor

# Crear el modelo con gradiente descendiente
modelo_sgd = SGDRegressor(max_iter=1000, tol=1e-3, learning_rate='optimal', penalty=None, random_state=42)

# Entrenar el modelo
modelo_sgd.fit(X_train, y_train)




```

```python
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Escalar datos para evitar problemas numéricos
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 🔹 Ajustar modelo con hiperparámetros mejorados
modelo_sgd = SGDRegressor(
    max_iter=5000,         # Más iteraciones para mejor convergencia
    tol=1e-4,             # Tolerancia más baja para asegurar precisión
    learning_rate='adaptive',  # Tasa de aprendizaje que se ajusta automáticamente
    eta0=0.0001,           # Tasa de aprendizaje inicial más baja
    penalty=None,         # Sin regularización para comparación con OLS
    random_state=42
)
modelo_sgd.fit(X_train_scaled, y_train)

# 🔹 Obtener coeficientes
intercepto_sgd = modelo_sgd.intercept_[0]  # β₀
coeficientes_sgd = modelo_sgd.coef_  # Coeficientes β

# 🔹 Crear un DataFrame con los coeficientes corregidos
coef_df_sgd = pd.DataFrame({"Variable": ["Intercepto"] + list(X.columns), 
                            "Coeficiente": [intercepto_sgd] + list(coeficientes_sgd)})

```

```python
y_pred = modelo_sgd.predict(X_test_scaled)
residuos = y_test - y_pred
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# 🔹 Predicciones en entrenamiento y test
y_train_pred = modelo_sgd.predict(X_train_scaled)
y_test_pred = modelo_sgd.predict(X_test_scaled)

# 🔹 Calcular métricas para entrenamiento
mse_train = mean_squared_error(y_train, y_train_pred)
rmse_train = np.sqrt(mse_train)
mae_train = mean_absolute_error(y_train, y_train_pred)
r2_train = r2_score(y_train, y_train_pred)

# 🔹 Calcular métricas para test
mse_test = mean_squared_error(y_test, y_test_pred)
rmse_test = np.sqrt(mse_test)
mae_test = mean_absolute_error(y_test, y_test_pred)
r2_test = r2_score(y_test, y_test_pred)

# 🔹 Mostrar resultados en un DataFrame
metrics_df = pd.DataFrame({
    "Métrica": ["MSE", "RMSE", "MAE", "R² Score"],
    "Entrenamiento": [mse_train, rmse_train, mae_train, r2_train],
    "Test": [mse_test, rmse_test, mae_test, r2_test]
})
```
