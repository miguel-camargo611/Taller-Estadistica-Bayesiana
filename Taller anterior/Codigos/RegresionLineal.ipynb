# Regresión Lineal



En este cuaderno se aborda el modelo de regresión lineal.  La estimación de los parámetros se realiza usando las herremiantas bayesiana.  Los problemas clave del modelado estadístico se pueden establecer en dos, primero definir las variables $x$ e $y$ (posiblemente usando transformaciones) de modo que la expectativa condicional de $y$ sea razonablemente lineal como una función de las columnas de $X$ con errores aproximadamente normales, y segundo establecer una distribución previa en los parámetros del modelo que refleje con precisión el conocimiento sustantivo, una distribución previa que sea lo suficientemente fuerte para que los parámetros del modelo se estimen con precisión a partir de los datos disponibles, pero no tan fuerte como para dominar los datos de manera inapropiada. 


En un modelo lineal múltiple se asume que $Y$ es una función lineal de sus predictores mas un ruido blanco: 


$$Y = \beta_{0} + \beta_{1} X_{1} + \beta_{2} X_{2} + \ldots \beta_{p} X_{p} + \epsilon $$


Los supuestos de un modelo lineal tradicional son:

* Todos los parámetros son lineales 
* La muestra de $n$ observaciones sigue el modelo poblacional del primer supuesto
* No existe perfecta colinealidad
* El error tiene un valor esperado de cero dado por variables independientes.
* Homocedasticidad: El error tiene la misma varianza que los valores de las variables explicativas.
* Heterocedasticidad: El error tiene una varianza diferente que los valores de las variables explicativas.

No hacemos suposiciones sobre las distribuciones (marginales o conjuntas) de
$ X_{i} $, pero suponemos que $ E [\epsilon | X] = 0 $, $ V[\epsilon | X] =
\sigma^{2} $, y ese $ \epsilon $ no está correlacionado entre las mediciones. La forma matricial del modelo es:



$$\mathbf{Y} = \mathbf{X}\mathbf{\beta} + \mathbf{\epsilon} $$

Donde $\mathbf{X}$ incluye una columna inicial de 1.

Cuando se incluye se asume que este debe tener un comportamiento normal  como se ve en la siguiente ecuación:

$$\mathbf{\epsilon} \sim MVN(\mathbf{0}, \sigma^2 \mathbf{I}) $$

independiente de $\mathbf{X}$.



Los coeficientes se estiman de la siguiente forma:



$$\widehat{\mathbf{\beta}} = (\mathbf{X}^T\mathbf{X})^{-1}\mathbf{X}^T\mathbf{Y}$$


Bajo el supuesto de ruido blanco, este es el mismo resultado que la estimación por máxima verosimilitud



$$\widehat{\mathbf{Y}} = \mathbf{X}\widehat{\mathbf{\beta}} = \mathbf{H}\mathbf{Y}$$


Donde $\mathbf{H}= \mathbf{X}(\mathbf{X}^T\mathbf{X})^{-1}\mathbf{X}^T$

Los residuales se obtienen son el resultado de la diferencia entre las observaciones y la estimación del modelo:


$$ SSE = \mathbf{Y}^{T} (\mathbf{I}- \mathbf{H})\mathbf{Y} $$

donde $\mathbf{I}-\mathbf{H}$ es simétrico e idempotente



***Cálculo de los estimadores.***


$$\widehat{\mathbf{\beta}}_{OLS}  =   argmin_{\beta \in \mathbb{R}^{2}} (\mathbf{Y}-\mathbf{X}\beta)^{T}(\mathbf{Y}-\mathbf{X}\beta)$$


$$\frac{\partial}{\partial \beta} =  (\mathbf{Y}^T\mathbf{Y}-2\mathbf{Y}^T(\mathbf{X}\mathbf{\beta} +\beta^{T} \mathbf{X}^{T}\mathbf{X}\beta))$$



$$\frac{\partial}{\partial \beta} = -2\mathbf{X}^{T}\mathbf{Y}+2\mathbf{X}^{T}\mathbf{X}\hat{\beta}=0$$



$$\hat{\beta}=(\mathbf{X}^{T}\mathbf{X})^{-1}\mathbf{X}^{T}\mathbf{Y}$$

El valor esperado y la varianza de llos estimadores es:

$$E[\hat{\beta}|x]=\beta$$

$$V[\hat{\beta}\mid x]=\sigma^{2}(\mathbf{X}^{T}\mathbf{X})^{-1}$$


A este procedimiento se le llama estimación de mínimos cuadrados ordinarios OLS en ingles y prueba que los errores al cuadrado se pueden minimizar obteniendo el valor de cada uno de los parámetros $\beta$.

***Tipos de variable***

| Y                     | X                      |
|-----------------------|------------------------|
| Variable dependiente  | Variable independiente |
| Variable explicada    |  Variable Explicatoria |
| Variable respuesta    |  Variable Control      |
| Variable predictiva   | Variable predictoria   |
| Regresada             |  Regresora             |

***Interpretación de los coeficiente***

| ***Intepretación del $\beta$*** | Dependiente | Independiente |                                         |
|:----------------------------------:|:-----------:|---------------|-----------------------------------------|
| Nivel-Nivel                        | $y$         | $x$           | $\Delta y = \beta_{i} \Delta x$         |
| Nivel-Log                          | $y$         | $log(x)$      | $\Delta y = \beta_{i} \Delta x$         |
| Log-Nivel                          |  $log(y)$   |  $x$          |  $\% \Delta y =(100 \beta_{i})\Delta x$ |
| Log-Log                            |  $log(y)$   |  $log(x)$     |  $\%\Delta y=\beta_{i} \% \Delta x$     |



El valor esperado de una función lineal condicionado al valor de los parámtreos es:


$$E(y_{i}\mid \beta, X)= \beta_{1}x_{i,1}+\beta_{2}x_{i,2}+\ldots+\beta_{p}x_{i,p}$$


Con el que se espera obtener la variación promedio de la variable dependiente $y$ de acuerdo a los parámetros que la relacionan con cada  $x_{p}$.




## Modelo Bayesiano

En el punto de vista bayesiano, formulamos una regresión lineal utilizando distribuciones de probabilidad en lugar de estimaciones puntuales. La variable $y$, no se estima como un valor único, sino que se supone que se extrae de una distribución de probabilidad. El modelo de Regresión Lineal Bayesiana con la respuesta muestreada de una distribución normal es:

$$y\sim N(\beta^{T}X,\sigma^{2}I)$$


Por lo tanto, es una estimación de la probabilidad de una variable independiente $\mathbf{X}$ dada un vector de parámetros $\psi$,  $p(\mathbf{X}\mid \psi)$ que se ajusta a las observaciones a través de la función likelihood conjunta $p(\mathbf{X},y \mid \psi. \theta)$. En el que la distribución previa $p(\psi, \theta)$ tiene la información para obtener los coeficientes de la regresión.

El resultado se genera a partir de una distribución normal (gaussiana) caracterizada por una media y una varianza. La media de la regresión lineal es la transposición de la matriz de ponderación multiplicada por la matriz predictora. La varianza es el cuadrado de la desviación estándar σ (multiplicada por la matriz de identidad porque esta es una formulación multidimensional del modelo).

El cambio entre un modelo lineal tradicional y uno bayesiano se aprecia en la información previa que existe en las observaciones y en la relación entre las mismas. Mientras que en la tradicional la relación independiente de $\theta$ se asume, en el modelo de regresión bayesiano esta se ignora. Es decir $p(\psi, \theta)=p(\psi)p(\theta)$.


Por lo tanto la distribución pesterior de los factores es:


$$p(\psi,\theta \mid \mathbf{X},y)=p(\sigma \mid \mathbf{X}) p(\theta \mid y)$$

Dado que el vector $\theta$ representa los parámetros del modelo, este se puede obtener como:

$$p(\theta \mid \mathbf{X},y)= p(\theta)p(y \mid \mathbf{X},\theta)$$


***Estimación de los parámetros*** 

Usando una distribución previa semiconjugada.  Se puede obtener $y\sim N(\mathbf{X}\mathbf{\beta} \Sigma_{y})$ donde $\mathbf{\beta}$ es el vector de parámetros estimado.

$$p(\mathbf{y} \mid X, \beta, \sigma^2)\propto exp\{-\frac{1}{2}SSE(\mathbf{\beta})\}$$

$$=exp\{-\frac{1}{2}(\mathbf{y}^T\mathbf{y}-2\mathbf{\beta}^{T}\mathbf{X}^T\mathbf{y})+\mathbf{\beta}^T\mathbf{X}^{T}\mathbf{X}\mathbf{\beta})$$


En donde $\mathbf{y}$ es una distribución multivariada normal y por lo tanto la distribución previa de $\mathbf{\beta}$ es también normal multivariada $(\mathbf{\beta}_{0}, \Sigma_{0})$.


$$p(\mathbf{\beta} \mid \mathbf{X},\mathbf{y},\sigma^{2}) $$

$$\propto p(\mathbf{y},\mathbf{\beta},\sigma^{2})p(\mathbf{\beta})$$


$$\propto exp\{-\frac{1}{2}(-2\mathbf{\beta}^{T}\mathbf{X}^T\mathbf{y}/\sigma^{2})-\frac{1}{2}(-2\mathbf{\beta}^{T}\Sigma_{0}^{-1}\mathbf{\beta}_{0}+\mathbf{\beta}^{T}\Sigma^{-1}_{0}\mathbf{\beta})\} $$

$$=exp\{\mathbf{\beta}^{T}\Sigma_{0}^{-1}\mathbf{\beta}_{0}+\mathbf{X}^{T}\sigma^{2})-\frac{1}{2}\mathbf{\beta}^{T}(\Sigma_{0}^{-1}+\mathbf{X}^{T}\mathbf{X}/\sigma^{2})\} $$


Obteniendo el valor esperado y la varianza se ve que si los elementos de la información previa de la matriz de covarianza $\Sigma_{0}^{-1}$, se obtiene el mismo resultado que en el modelo tradicional. Por otra parte si $\sigma^{2}$ es grande, implica que la presición es menor y la información previa tiene una papel mayor en la estimación de los coeficientes.


$$Var [\mathbf{\beta} \mid \mathbf{y}, \mathbf{X}, \sigma^{2}] = (\Sigma_{0}^{-1}+ \mathbf{X}^{T}\mathbf{X}/\sigma^{2})^{-1} $$

$$E[\mathbf{\beta} \mid \mathbf{y},\mathbf{X},\sigma^{2}] = (\Sigma_{0}^{-1}+ \mathbf{X}^{T}\mathbf{X}/\sigma^{2})^{-1}( \Sigma_{0}^{-1}+ \mathbf{X}^{T}\mathbf{y}/\sigma^{2}) $$

De igual forma que en la distribución normal, $\sigma^{2}$ se estima con ayuda de una inverse-gamma en donde $\gamma =1/\sigma^{2}$ es la presición y se obtiene como $\gamma \sim gamma(\nu_{0}/2, \nu_{0}\sigma_{0}^{2}/2)$.


$$p(\gamma \mid \mathbf{y},\mathbf{X},\mathbf{\beta}) \propto p(\gamma) p(\mathbf{y} \mid \mathbf{X},\mathbf{\beta}, \gamma) $$

$$ \propto [\gamma^{\nu_{0}/2 -1}exp(-\gamma \times \nu_{0}\sigma_{0}^{2}/2)] \times[\gamma^{\nu_{0}/2}exp(-\gamma \times SSE(\mathbf{\beta})/2)] $$

Esto siginifica que la varianza se obtiene s partir de una distribución inverse-gamma.

$$\{ \sigma^{2} \mid \mathbf{y}, \mathbf{X},\mathbf{\beta} \} inverse-gamma([\nu_{0}+n]/2, [\nu_{0}\sigma_{0}^{2}+SSE(\beta)]/2) $$


Usando el muestre de Gibbs se puede estimar el valor de los coeficientes de la regresión realizando los siguientes pasos:

Dados los valores que se generan en la cadena de markov de $ \{ \beta^{s}, \sigma^{2(s)} \}$ se obtiene el valor de $\beta^{s+1}$ y $\sigma^{2(s+1)}$ de la siguiente forma:


* Actualiza $\mathbf{\beta}$ 

    a. Calcula $\mathbf{V}= Var[\mathbf{\beta} \mid \mathbf{y}, \mathbf{X}, \sigma^{2(s)}]$ y $\mathbf{m}=E[\mathbf{\beta} \mid \mathbf{y},\mathbf{X},\sigma^{2(s)}]$

    b. Muestra de $\mathbf{\beta}^{(s+1)} \sim multivariate-normal(\mathbf{m},\mathbf{V})$ 

* Actualiza $\sigma^{2(s+1)}$ 

    a. Calcula $SSE(\mathbf{\beta}^{(s+1)})$

    b. Muestra de $\sigma^{2(s+1)} \sim inverse-gamma(\nu_{0}/2, \nu_{0}\sigma_{0}^{2}/2)$


A veces, se debe realizar un análisis en ausencia de información previa precisa, o información que se pueda convertir fácilmente en los parámetros de una distribución previa conjugada. En estas situaciones, uno podría ceñirse a la estimación por mínimos cuadrados, con el inconveniente de que las declaraciones de probabilidad sobre $\beta$ no estarían disponibles. Alternativamente, a veces es posible justificar una distribución anterior con otros criterios. Una idea es que, si la distribución previa no va a representar información real sobre los parámetros, entonces debería ser lo menos informativa posible. 

El resultado es una distribución posterior resultante de alguien que comenzó con poco conocimiento de la población que se estudia. En estos casos se implementa una distribución uniforme previa que centra el resultado de los coeficientes alrededor de el resultado que se obtiene por estimación de mínimos cuadrados ordinarios $\beta=\beta_{OLS}$.

En este caso la covarianza $\Sigma_{0}^{-1}=(\mathbf{X}^{T}\mathbf{X})/ (n \sigma^{2}) $ que en cierto caso usa información de $\mathbf{y}$ para crearse. 

### Primer Ejemplo

En etsa primera parte usamos dos vectores creados al azar y generamos los coeficientes de la regresión, usando primero mínimos cuadrados ordinarios y luego la distribución de Gibbs. que se se menció en la sección anterior.

```python
import pymc as pm
import numpy as np
import arviz as az

# Datos simulados
np.random.seed(42)
X = np.random.randn(100, 2)
beta_true = [1.0, 2.5]
y = X @ beta_true + np.random.randn(100) * 0.5

with pm.Model() as model:
    # Priors para los coeficientes
    beta = pm.Normal('beta', mu=1, sigma=1, shape=2)
    
    # Prior para la desviación estándar
    sigma = pm.HalfCauchy('sigma', beta=1)
    
    # Modelo lineal
    mu = pm.math.dot(X, beta)
    
    # Likelihood
    y_obs = pm.Normal('y_obs', mu=mu, sigma=sigma, observed=y)
    
    # Inferencia
    trace = pm.sample(10000, return_inferencedata=True)
    
# Resumen de los resultados
az.summary(trace)
```

```python
# Plot de las distribuciones posteriores
az.plot_posterior(trace)
plt.show()
```

```python
# Revisar las divergencias
divergences = trace.sample_stats.diverging.sum().values
print(f"Total divergences: {divergences}")
```

```python
import pymc as pm
import numpy as np
import arviz as az
import matplotlib.pyplot as plt

# Datos simulados: predictores con relación débil a la respuesta
np.random.seed(42)
X = np.random.randn(100, 2)  # Dos predictores
beta_true = [0.01, 0.01]  # Coeficientes reales muy pequeños
y = X @ beta_true + np.random.randn(100) * 5  # Agregar mucho ruido

with pm.Model() as model:
    # Priors para los coeficientes
    beta = pm.Normal('beta', mu=0, sigma=10, shape=2)
    
    # Prior para la desviación estándar
    sigma = pm.HalfCauchy('sigma', beta=1)
    
    # Modelo lineal
    mu = pm.math.dot(X, beta)
    
    # Likelihood
    y_obs = pm.Normal('y_obs', mu=mu, sigma=sigma, observed=y)
    
    # Inferencia con más cadenas y pasos para intentar alcanzar convergencia
    trace = pm.sample(2000, chains=4, return_inferencedata=True, target_accept=0.9)

# Resumen de los resultados
az.summary(trace)

# Plot de las distribuciones posteriores
az.plot_posterior(trace)
plt.show()
```

# 📝 Checklist para Modelos Bayesianos usando MCMC

## 📦 Preparación de Datos
- **Revisar los datos**: Asegúrate de que los datos estén limpios, sin valores faltantes ni errores.
- **Genera Datos Simulados**: Crea una simulación de la base usando parámetros artificiales para desarrollar un modelo que estime esos parámetros.
- **Exploración de datos**: Realiza un análisis exploratorio de los datos para entender su distribución y relaciones.

## 📋 Definición del Modelo
-  **Especificar el modelo bayesiano**: Define la estructura del modelo, incluyendo:
  - Variables observadas.
  - Variables latentes (si existen).
  - Relaciones funcionales entre las variables.
- **Seleccionar los priors**: Elige distribuciones a priori adecuadas para los parámetros:
  - Priors no informativos para casos con poca información previa.
  - Priors informativos si se dispone de conocimiento previo.
  - Evita priors excesivamente informativos a menos que esté justificado.
- **Verificar la suposición de los priors**: Visualiza los priors para verificar que se ajusten a las creencias previas.

## 🛠 Construcción del Modelo en el Software
- **Implementar el modelo**: Usa librerías como `PyMC`, `Stan`, `NumPyro`, etc., para construir el modelo.
- **Incluir la verosimilitud (likelihood)**: Asegúrate de que la verosimilitud esté correctamente especificada en función de los datos observados.

## 🚀 Muestreo MCMC
- **Simulación del Modelo**: Usa los datos previamente simulados y asegurate que obtienes los parámetros que usaste para simularlos.
- **Seleccionar el algoritmo MCMC**: Elige el algoritmo adecuado (`NUTS`, `Metropolis-Hastings`, `HMC`, etc.).
- **Establecer parámetros de muestreo**:
  - Número de cadenas (`chains`).
  - Número de muestras (`draws`).
  - Período de adaptación (`tune`).
- **Establecer criterios de aceptación**: Ajusta el `target_accept` para mejorar la exploración.
- **Monitorear advertencias**: Verifica si hay advertencias durante el muestreo, como muestras divergentes o problemas de exploración.

## 📊 Evaluación del Modelo
- **Graficar las trazas**: Visualiza las trazas de las cadenas para inspeccionar la convergencia.
- **Calcular criterios de convergencia**:
  - **R-hat**: Cercano a 1 indica buena convergencia.
  - **Número efectivo de muestras (ESS)**: Evalúa la independencia de las muestras.
  - **Autocorrelación**: Analiza la autocorrelación en las cadenas.
- **Analizar divergencias**: Identifica muestras divergentes que puedan sugerir problemas en la especificación del modelo.
- **Inspeccionar las distribuciones posteriores**: Visualiza las distribuciones posteriores para los parámetros clave.
  
## 🧮 Comparación y Selección de Modelos
- **Calcular métricas de comparación**:
  - WAIC (Watanabe-Akaike Information Criterion).
  - LOO (Leave-One-Out Cross-Validation).
- **Calcular el Factor de Bayes**: Si se requiere comparar modelos y es factible.
- **Realizar comprobaciones de ajuste**: Usar posterior predictivo (`posterior predictive checks`) para evaluar el ajuste del modelo a los datos observados.

## 📈 Ajuste y Validación
- **Revisar el ajuste**: Si el modelo no se ajusta bien, revisa las siguientes opciones:
  - Ajustar priors.
  - Incluir más datos o predictores.
  - Cambiar la estructura del modelo.
- **Validar el modelo**: Divide los datos en conjuntos de entrenamiento y prueba, si es aplicable.

### Segundo Ejemplo 

En el siguiente ejemplo se crearán dos variables independientes $x_1$ y $x_2$ y una dependiente $y$. Primero se estimarán los coeficientes $\beta$ usando el método de mínimos cuadrados. Luego se estimarán los coeficientes $\beta$ usando el método bayesiano aplicando el muestreo de Gibbs.

Para este caso se usará la base de datos de la felicidad que contiene la información ajustada de los países más felices en el mundo de acuerdo a unas variables sociales y económicas que se re escalan para dar un puntaje a cada país.


Las variables son:

* Country or region = Países
* Score = Puntaje de felicidad
* GDP per capita = Pib per capita 
* Social support = Ayuda social
* Healthy life expectancy = Expectativa de vida
* Freedom to make life choices = Libertad para hacer elecciones
* Generosity = Generosidad
* Perceptions of corruption = Percepción de la corrupción.





```python
import pandas as pd

# Descargar y cargar los datos
url = "https://raw.githubusercontent.com/Andres1984/Estadistica2/master/Codigos/2019.csv"
df = pd.read_csv(url, index_col=0)

# Seleccionar las variables para el modelo (sin escalar)
df = df[['Score', 'GDP per capita', 'Social support', 'Healthy life expectancy', 
         'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']]

# Definir variables independientes y la variable objetivo
X = df.drop(columns='Score').values
y = df['Score'].values

```

```python
df
```

```python
import pymc as pm
import numpy as np
import arviz as az
import matplotlib.pyplot as plt
# Obtener el número de predictores

# Obtener el número de predictores
n, p = X.shape

# Modelo Bayesiano sin escalar las variables
with pm.Model() as modelo_bayesiano:
    # Priors para los coeficientes beta (no informativos)
    beta = pm.Normal('beta', mu=0, sigma=10, shape=p)
    
    # Prior para la desviación estándar
    
    sigma = pm.HalfCauchy('sigma', beta=1)
    
    # Modelo lineal
    mu = pm.math.dot(X, beta)
    
    # Likelihood
    y_obs = pm.Normal('y_obs', mu=mu, sigma=sigma, observed=y)
    
    # Muestreo de Gibbs (MCMC)
    trace = pm.sample(2000, chains=4, target_accept=0.9, return_inferencedata=True)

    # Posterior Predictive Check
    posterior_pred = pm.sample_posterior_predictive(trace, var_names=["y_obs"], return_inferencedata=True)

```

```python
pm.model_to_graphviz(modelo_bayesiano)
```

```python
import matplotlib.pyplot as plt

# Graficar las trazas
az.plot_trace(trace)
plt.show()

# Calcular criterios de convergencia
summary = az.summary(trace)
print(summary)

# Verificar divergencias
divergences = trace.sample_stats.diverging.sum().values
print(f"Total divergences: {divergences}")

# Graficar distribuciones posteriores
az.plot_posterior(trace)
plt.show()
```

```python


# Usar directamente el objeto `posterior_pred` como `idata`
idata = posterior_pred

# Extraer las predicciones y calcular estadísticas
mean_preds = idata.posterior_predictive["y_obs"].mean(dim=["chain", "draw"])
lower_preds = idata.posterior_predictive["y_obs"].quantile(0.025, dim=["chain", "draw"])
upper_preds = idata.posterior_predictive["y_obs"].quantile(0.975, dim=["chain", "draw"])

# Graficar los resultados
plt.figure(figsize=(10, 6))
plt.plot(y, 'b.', label='Valores Reales')
plt.plot(mean_preds, 'r.', label='Predicciones Media')
plt.fill_between(range(len(y)), lower_preds, upper_preds, color='gray', alpha=0.5, label='Intervalo de 95%')
plt.xlabel('Índice de País')
plt.ylabel('Score de Felicidad')
plt.legend()
plt.title('Predicciones del Modelo Bayesiano')
plt.show()


```

### Método de Estimación por mínimos cuadrados

Usando la variables "score" que representa el índice de la felicidad como variable dependiente, se genera un modelo de regresión con respecto a las demás variables independientes.

### Método de Estimación usando inferencia Bayesiana.

Ahora usaremos la metodología explicada al inicio de este cuaderno para obeter los coeficientes de la regresión con ayuda de un proceso de remuestreo de Gibbs. Este debería disminuir el error de estimación. Esto se obtiene gracias a que se usan diferentes muestras y se encuentra un estimador promedio que se disminuye el error de la regresión. 

En este proceso de estimación, los coeficientes provienen del resultado de una distribución uniforme previa que se  multiplica por la distribución de la variable dependiente para obeter una distribución posterior.
